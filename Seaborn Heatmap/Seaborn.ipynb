```python
seaborndata1 = pd.read_csv("../Output_data/Boise1.csv")

seaborndata1.head()
```

### BOISE DATA 

```python
sns.set()
plt.figure(figsize=(10,10))
ax = sns.heatmap(seaborndata1.set_index('Year').T, vmin=-10, vmax= 30, cmap ="RdBu_r")
ax.set_title('Boise Mean temperature every month from 1951 to 2017')
plt.savefig("../Images/Boise Seaborn.png")
plt.show()
```
