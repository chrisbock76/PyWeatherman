## Monthly Statistics

* Create a list of elements corresponding to each month over the entire range of years + months


* Run One-sample t-test on the change in temperatures from initial value


* Display the summary results


* Loop over all months

```python
months_df=['JANUARY','FEBRUARY','MARCH','APRIL','MAY','JUNE','JULY','AUGUST','SEPTEMBER','OCTOBER','NOVEMBER','DECEMBER']

for column in df.columns[1:]:
#     print(df[column])
#     print(df[column].name)
    print(f'{df[column].name} Monthly Statistics')
    print('------------------------')
    data_city = (df[column])
    
    imonth = 0 
    for month in months_df:
        data=data_city[imonth::12] 
        data = data.reset_index(drop=True)
        data = data-data[0]
        j = stats.ttest_1samp(data, 0) 
        p_val=j[1]/2
              
        if p_val < 0.05:
            if data.mean(axis = 0) > 0 :
                print("%s : Significant warming found ; Mean deltaT: % 2.1f, p-val: % 1.5f" %(month, data.mean(axis = 0)  , p_val)) 
            else :
                print("%s : Significant cooling found ; Mean deltaT: % 2.1f, p-val: % 1.5f" %(month, data.mean(axis = 0)  , p_val))
        else :
            print("%s : NO Significant warming/cooling found; Mean deltaT: % 2.1f, p-val: % 1.5f" %(month, data.mean(axis = 0), p_val))
        
        imonth = imonth + 1
    print('')
```
